```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wtl = np.load("val_loss_cnn_wtl.npy")
tl_1 = np.load("val_loss_cnn_tl(1).npy")
tl_2 = np.load("val_loss_cnn_tl(2).npy")
```

```python
total_data = 30
wtl = wtl[:total_data]
tl_1 = tl_1[:total_data]
tl_2 = tl_2[:total_data]
```

```python
dashes = [(None, None), [2, 2], [7, 4, 3, 4]]

line, = plt.plot(wtl, label="Tanpa Transfer Learning")
line.set_dashes(dashes[0])

line, = plt.plot(tl_1, label="Transfer Learning Skenario (1)")
line.set_dashes(dashes[1])

line, = plt.plot(tl_2, label="Transfer Learning Skenario (2)")
line.set_dashes(dashes[2])

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Values for Each Epochs")
plt.legend(handlelength=5)

plt.show()
```
